```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
price_df = pd.read_excel('PRICE INDEX.xlsx').set_index('Date')
price_df.dropna(inplace=True)
```

```python
price_df
```

```python
snp = price_df[price_df.columns[0]].values
```

```python
dow = price_df[price_df.columns[7]].values
```

```python
dax = price_df[price_df.columns[2]].values
```

```python
topix = price_df[price_df.columns[3]].values
```

```python
fig  = plt.figure(figsize = (16,9))
plt.plot(price_df.index , snp/sum(snp), label = 'S&P500')
plt.plot(price_df.index ,dow/sum(dow), label = 'Dow Jones')
plt.plot(price_df.index ,dax/sum(dax), label = 'DAX')
plt.plot(price_df.index ,topix/sum(topix), label = 'TOPIX')
plt.legend()
plt.grid()
plt.savefig('fig3_11.png', dpi=300)
```
