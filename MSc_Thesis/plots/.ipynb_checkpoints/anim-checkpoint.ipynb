```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
import numpy as np
```

```python
orig_data_frame = pd.read_excel('PRICE INDEX.xlsx')
data_frame = orig_data_frame.dropna().set_index('Date')
monthly_prices = data_frame.resample("M").mean()
snp = pd.DataFrame( monthly_prices[monthly_prices.columns[0]] , monthly_prices.index )
# sns.lineplot(snp.index , snp[snp.columns[0]], data=snp , color='r')
```

```python
def animate(i):
    data = snp.iloc[:int(i+1)]
    x = data.index
    y = data[data.columns[0]]
    line.set_data(x,y)
    return line
    
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

ax.set_xlim(1999, 2016)
ax.set_ylim(np.min(snp[snp.columns[0]]), np.max(snp[snp.columns[0]]))
ax.set_xlabel('Year',fontsize=20)
ax.set_ylabel(snp.columns[0],fontsize=20)
plt.title('snp',fontsize=20)  
line, = ax.plot([], [])

Writer = animation.writers['ffmpeg']
writer = Writer( fps=20 , metadata=dict(artist='Me'))

ani = animation.FuncAnimation(fig, animate, frames=1000 , repeat=True , interval = 50)  
ani.save('snp.mp4', writer=writer)
```
