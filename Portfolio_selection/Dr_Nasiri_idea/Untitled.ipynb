```python
import numpy as np
import itertools as it
import matplotlib.pyplot as plt
```

```python
import os
direct_file_names = os.listdir()
npy_files = []
for i in direct_file_names:
    if i[-3:]=='npy':
        npy_files.append(i)
```

```python
eight_stocks=np.load('fitness_p_count50steps1000retain0.3rnd0.3mutate0.30.790761868454022.npy',allow_pickle=True)
eight_stocks_min = min(eight_stocks[1:,0])
```

```python
total_fitness_data = []
for i in range(len(npy_files)):
    total_fitness_data.append(np.load(npy_files[i] , allow_pickle=True)[1:])
```

```python
total_portfos = []
for k in range(8):
    total_portfos.append([])
er=[]
for i in range(len(npy_files)):
    total_portfos[len(total_fitness_data[i][0][1])].append(min(total_fitness_data[i][:,0]))
    er.append(len(total_fitness_data[i][0][1]))
    
```

```python
mean_tot_portfos = []
for i in range(2,8):
    mean_tot_portfos.append(np.mean(total_portfos[i]))
mean_tot_portfos.append(eight_stocks_min)    
```

```python
plt.plot(mean_tot_portfos,'o--')
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
fig, ax = plt.subplots(figsize=(16,9))
plt.plot(mean_tot_portfos,'o-', markersize=14)

plt.xlabel('Num of stocks in portfolio' , fontsize=15, fontweight='black', color = '#333F4B')
plt.ylabel('Optimized Portfolio Risk' , fontsize=15, fontweight='black', color = '#333F4B')

plt.xticks([0,1,2,3,4,5,6,7],[2,3,4,5,6,7,8] )
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
plt.savefig('topten_fig6.png', dpi=300, bbox_inches='tight' , transparent=False)
```
