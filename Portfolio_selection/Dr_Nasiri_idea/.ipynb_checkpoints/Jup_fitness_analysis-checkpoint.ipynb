```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
direct_file_names = os.listdir()
npy_files = []
for i in direct_file_names:
    if i[-3:]=='npy' and i[22]=='1' and i[34]=='3' and i[40]=='3'\
    and i[49]=='3' and i[15]=='5':
        npy_files.append(i)
```

```python
npy_files
```

```python
total_fitness_data = []
for i in range(len(npy_files)):
    total_fitness_data.append(np.load(npy_files[i] , allow_pickle=True)[1:])
```

```python
len(total_fitness_data[6][0][1])
min_risk_portfos = np.zeros(9)
for i in range(7):
    min_risk_portfos[len(total_fitness_data[i][0][1])] = min(total_fitness_data[i][:,0])
min_risk_portfos
```

```python
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
fig, ax = plt.subplots(figsize=(16,9))
plt.plot(min_risk_portfos[2:],'o-', markersize=14)

plt.xlabel('Num of stocks in portfolio' , fontsize=15, fontweight='black', color = '#333F4B')
plt.ylabel('Optimized Portfolio Risk' , fontsize=15, fontweight='black', color = '#333F4B')

plt.xticks([0,1,2,3,4,5,6,7],[2,3,4,5,6,7,8] )
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
plt.savefig('topten_fig6.png', dpi=300, bbox_inches='tight' , transparent=False)
```
