```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
direct_file_names = os.listdir()
npy_files = []
for i in direct_file_names:
    if i[-3:]=='npy' and i[22]=='1' and i[34]=='3' and i[40]=='3'\
    and i[49]=='3' and i[15]=='5':
        npy_files.append(i)
```

```python
total_fitness_data = []
for i in range(len(npy_files)):
    total_fitness_data.append(np.load(npy_files[i] , allow_pickle=True)[1:])
```

```python
one_final_fit =[]
for i in range(len(npy_files)):
    for j in range(len(total_fitness_data[i])):
        one_final_fit.append(total_fitness_data[i][j])
one_final_fit = np.array(one_final_fit)        
```

```python
fig= plt.figure(figsize=(16,9))
plt.plot(one_final_fit[:,0][one_final_fit[:,0]<0.073])
```

```python
lower_fitness = one_final_fit[one_final_fit[:,0]<0.075]
```

```python
port_min = lower_fitness[np.where(lower_fitness[:,0]==min(lower_fitness[:,0]))[0][0]][1]
port_max = lower_fitness[np.where(lower_fitness[:,0]==max(lower_fitness[:,0]))[0][0]][1]
```

```python
def diff_function(pop , ind):
    dif =[]
    for i in range(len(pop)):
        u=0
        for j in range(len(ind)):
             u+=(lower_fitness[i,1][j]-ind[j])**2
        dif.append(u)
    return np.std(dif)/2
```

```python
final_portfos = [port_min , port_max ]

for i in range(len(lower_fitness)):
    kk=0
    for j in range(len(final_portfos)):
        c=0
        for k in range(8):
            c+=(lower_fitness[i,1][k]-final_portfos[j][k])**2
        if c >= diff_function(lower_fitness[:,1] , final_portfos[j]):
            kk+=1
    if kk==len(final_portfos):
        final_portfos.append(lower_fitness[i,1])
        print('one portfo found')
```

```python
fig = plt.figure(figsize=(16,9))
plt.plot(final_portfos)
plt.show()
```
