# Peter Holland Quasi-Potential

```python
from my_packs import coupled_markets as cm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('PRICE INDEX.xlsx')
df = df.set_index('Date')
df = df.dropna()
```

```python
snp_ret = cm.price_return(df[df.columns[0]])
ftse_ret = cm.price_return(df[df.columns[1]])
```

```python
pdf = cm.joint_pdf(snp_ret,ftse_ret,0.001,0.001)
ker_2d = cm.kernel_2d(pdf['pdf'],3,3)
qpot = cm.quasi_qpot(ker_2d)
```

```python
fig = plt.figure(figsize=(16,9))
plt.imshow(abs(qpot['Q_tot']),origin='lower')
plt.colorbar()
```
