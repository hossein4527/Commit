# Working with scipy.sparse to get joint pdf

```python
import numpy as np
import scipy.sparse as sps
import pandas as pd
from my_packs import coupled_markets as cm
```

```python
df1 = pd.read_csv('data-s&p.csv')
df2 = pd.read_csv('data-dax.csv')
df1 = df1['price']
df2 = df2.price
d1 = cm.price_return(df1)
d2 = cm.price_return(df2)
d1_std = np.std(d1)
d2_std = np.std(d2)
d1 /= np.std(d1)
d2 /= np.std(d2)

# d1 = np.random.normal(size=1000000)

# d2 = np.random.normal(size=1000000)
delta = 0.2

up_limit1 = int(np.amax(d1)/delta+2)*delta
down_limit1 = int(np.amin(d1)/delta)*delta
num_bins1 = int((up_limit1 - down_limit1)/delta)

up_limit2 = int(np.amax(d2)/delta+2)*delta
down_limit2 = int(np.amin(d2)/delta)*delta
num_bins2 = int((up_limit2 - down_limit2)/delta)

bins1 = np.arange(down_limit1,up_limit1,delta)
bins2 = np.arange(down_limit2,up_limit2,delta)

cim1 = np.zeros(len(d1))
cim2 = np.zeros(len(d2))


for i in range(num_bins1-1):
    block = np.logical_and(d1>=bins1[i],d1<bins1[i+1])
    cim1[block] = i

for i in range(num_bins2-1):
    block = np.logical_and(d2>=bins2[i],d2<bins2[i+1])
    cim2[block] = i
    
j = sps.csr_matrix((np.ones_like(cim2),(cim2,cim1)),shape=(num_bins2,num_bins1))
pjoint = j.todense()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,9))
plt.imshow(pjoint)
plt.colorbar()
plt.xticks(range(num_bins1),np.around(bins1*d1_std,3))
plt.yticks(range(num_bins2),np.around(bins2*d2_std,3))
plt.show()
```

```python
pdf = cm.joint_pdf(d1,d2,0.2,0.2)['pdf']
```
