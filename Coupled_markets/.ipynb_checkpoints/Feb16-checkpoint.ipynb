# Invetigating around Mass in markets

### 1. The first approach is to look at the joint pdf and find a definition for mass:

Here we call the total number of shares of a market at time "t", $m_{tot}$ and the trading volume of the market at the same time, $\delta m$. So one may refer to the mobility of a market as the ratio $\frac{\delta m}{m_{tot}}$, while $\delta m$ is just our good old trading volume the market, we need to find the data covering the $m_{tot}$ part meaning the total number of shares of the mentioned market.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### S&P500

```python
df1 = pd.read_csv('data-s&p.csv')
delta_m1 = df1['volume']
```

```python
plt.plot(delta_m1,'.-')
```

### Dow Jones

```python
df2 = pd.read_csv('data-dow.csv')
delta_m2 = df2['volume']
```

```python
plt.plot(delta_m2,'.')
```

```python
fig , (ax1,ax2) = plt.subplots(ncols=2,figsize=(16,9))
ax1.plot(delta_m1)
ax1.plot(delta_m2)
ax2.plot(delta_m1/900,'.')
ax2.plot(delta_m2/100,'.')
```

```python
newdfsp = pd.read_csv('GSPC.csv')
newdfdj = pd.read_csv('DJI.csv')
# newdfsp[400:500]
```

```python
fig , (ax1,ax2) =  plt.subplots(ncols=2,figsize=(21,8))

ax1.plot(np.linspace(0,len(newdfsp['Volume'][420:]),len(newdfsp['Volume'][420:])),newdfsp['Volume'][420:])
ax1.set_title('S&P500',fontsize=20)
ax1.set_xlabel('Date',fontsize=18)
ax1.set_ylabel('Volume',fontsize=19)

ax2.plot(newdfdj['Volume'])
ax2.set_title('Dow Jones',fontsize=20)
ax2.set_xlabel('Date',fontsize=18)
ax2.set_ylabel('Volume',fontsize=18)

plt.show()
```

```python
dow_shares = 82
dow_vol = 3.8
dow_cap = 21

snp_shares = 934
snp_vol = 87.3
snp_cap = 259

rus_shares = 280
rus_vol = 22.2
rus_cap = 43
```

```python
dow_deltam_m = dow_vol/dow_shares
snp_deltam_m = snp_vol/snp_shares
rus_deltam_m = rus_vol/rus_shares
```

```python
fig = plt.figure(figsize=(16,9))
plt.plot([dow_cap,snp_cap,rus_cap],[dow_deltam_m,snp_deltam_m,rus_deltam_m],'o')
plt.ylabel('delta m/m',fontsize=18)
```
